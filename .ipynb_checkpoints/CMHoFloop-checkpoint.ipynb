```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#get names of hall of fame
start_url = 'https://countrymusichalloffame.org/hall-of-fame/members/'
listurl = [start_url]

for x in range(2,17):
    url = start_url + 'page/' + str(x)
    listurl.append(url)
listurl
```

```python
#save raw html
responses = []
for url in listurl:
    print(url)
    response = requests.get(url)
    print(response.status_code)
    responses.append(response)
```

```python
#loop through raw responses
allmembers = pd.DataFrame()
for htm in responses:
    #make soup object
    soup = BeautifulSoup(htm.content, 'html.parser')
    #locate section of page needed
    pagegroup = soup.find_all('div', attrs = {'class': 'vertical-card'})
    #empty list to save work
    inductyear = []
    inductname = []
    inductlink = []
    for x in pagegroup:
        pos = 0
        #loop to save name and year in known postions
        for string in x.stripped_strings:
            if pos == 0:
                inductname.append(string)
            elif pos == 1:
                inductyear.append(string[-4:])
            pos += 1
            print(repr(string))
        h = 0
        #loop to extract link to details page
        for child in x.children:
            if h == 3:
                link = child.find_all('a')[0].get('href')
                print(link)
                inductlink.append(link)
            h += 1
    #build as dataframe
    members = pd.DataFrame(inductname, columns=['name'])
    members['Inducted'] = inductyear
    members['link'] = inductlink
    allmembers = allmembers.append(members)
```

```python
#save output
allmembers.reset_index(drop=True)
allmembers.to_csv('allmembers.csv', index=False)
```

```python
#get details on each link
responses = []
x= 0
for url in allmembers['link']:
    print(url)
    response = requests.get(url)
    print(response.status_code)
    x+=1
    print(x)
    responses.append(response)
```

```python
memberdetailslist = []
for response in responses:
    soupmd = BeautifulSoup(response.content, 'html.parser')
    memberdetails = soupmd.find_all('div', attrs = {'class': 'fl-rich-text detail-wrapper'})
    memberdetailslist.append(memberdetails)
```

```python
#gather details
memdict = {}
name = ''
for x in memberdetailslist:
    pos = 0
    for string in x[0].stripped_strings:
        if string.startswith('Birthplace'):
            memdict[name].append(string[12:].strip())
        elif string.startswith('Birth'):
            memdict[name].append(string.strip())
        elif pos == 1:
            memdict[string] = []
            name = string
        pos += 1
        print(str(pos) + " " + string[0:9])
memdict
```

```python
memberdetaildf = pd.DataFrame.from_dict(memdict, orient='index')
memberdetaildf = memberdetaildf.reset_index().rename(columns={"index": "name",0:'life',1:'birthplace'})
split = memberdetaildf.life.str.split('-',expand = True)
split['birthdate'] = split[0].apply(lambda x: str(x).strip()[7:])
split['death'] = split[1].apply(lambda x: str(x).strip()[7:])
memberdetaildf['birthdate'] = split['birthdate']
memberdetaildf['death'] = split['death']
memberdetaildf.drop(columns=['life'], inplace=True)
memberdetaildf['Inducted'] = allmembers['Inducted']
memberdetaildf
```

```python
memberdetaildf.to_csv('memberdetail.csv', index=False)
```
