```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start_url = 'https://countrymusichalloffame.org/hall-of-fame/members/'
listurl = [start_url]

for x in range(2,17):
    url = start_url + 'page/' + str(x)
    listurl.append(url)
listurl
```

```python
responses = []
for url in listurl:
    print(url)
    response = requests.get(url)
    print(response.status_code)
    responses.append(response)
```

```python
allmembers = pd.DataFrame()
for htm in responses:
    soup = BeautifulSoup(htm.content, 'html.parser')
    pagegroup = soup.find_all('div', attrs = {'class': 'vertical-card'})
    inductyear = []
    inductname = []
    inductlink = []
    for x in pagegroup:
        pos = 0
        for string in x.stripped_strings:
            if pos == 0:
                inductname.append(string)
            elif pos == 1:
                inductyear.append(string[-4:])
            pos += 1
            print(repr(string))
        h = 0
        for child in x.children:
            if h == 3:
                link = child.find_all('a')[0].get('href')
                print(link)
                inductlink.append(link)
            h += 1
    members = pd.DataFrame(inductname, columns=['name'])
    members['Inducted'] = inductyear
    members['link'] = inductlink
    allmembers = allmembers.append(members)
```

```python
allmembers.reset_index(drop=True)
allmembers.to_csv('allmembers.csv', index=False)
```
